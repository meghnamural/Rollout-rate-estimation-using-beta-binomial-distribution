VARIABLES:     alpha,b- shape parameters
               t- time period ( in months)


```
import math
from matplotlib import pyplot as plt
from scipy.optimize import minimize
from scipy.stats import beta
import random
import numpy as np
import pandas as pd
import scipy.special
from scipy.special import gammaln
```

Function:  probability() 	 : Calculates the response rate to a mailing

```
def probability(alpha,b,ms,xs):
	p=[]
	for i in range(0,126):
		m=ms.iloc[i]
		x=xs.iloc[i]
		pt = gammaln(m+1) + gammaln(x+alpha) + gammaln(m-x+b) + gammaln(alpha+b) - \
		(gammaln(x+1) + gammaln(m-x+1) + gammaln(alpha) + gammaln(b) + gammaln(m+alpha+b))
		p=np.append(p,pt)
	return p	
```


Function: distribution()      		: Calculates the probability distribution

```
def distribution(alpha,b,ms,xs):
	g=[]
	for i in range(0,126):

		m=ms.iloc[i]
		x=xs.iloc[i]
		g=beta.pdf(i,alpha,b)
		print(g)
	return g
```

Function:   maximize()               : Maximizes the log likelihood to determine the optimal model parameters   

```
def maximize(ms,xs):
	func = lambda x: -log_likelihood(x[0], x[1], ms, xs)
	x0 = np.array([.1, .1])
	r = minimize(func, x0, method='Nelder-Mead', options={'disp': True})
	return r
```


Function:   log_likelihood()			: Calculates the log likelihood of the BBD model parameters

```
def log_likelihood(alpha,b,ms,xs):
	prob = probability(alpha,b,ms,xs)
	final_log_likelihood=np.sum(prob)
	return(final_log_likelihood)	
```
